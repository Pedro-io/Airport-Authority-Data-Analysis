# Análise Inicial do Dados
Este notebook tem como objetivo a análise do conjunto de dados que será utilizado no projeto 

```python
# importando bibliotecas que seram utilizadas
import pandas as pd 
import os 
```

## importando arquivos que seram utilizados 

```python
# Pasta onde as planilhas estão salvas
pasta_planilhas = "C:\\Users\\maype\\Desktop\\projetos\\Airport Authority Data Analysis\\Data" # Caminho das pastas 

# Lista para guardar os dataframes das planilhas 
dataframes = []

# Laço para iterar sobre os arquivos da pasta buscando todos os arquivos csv

for arquivo in os.listdir(pasta_planilhas): 
    if arquivo.endswith(".csv"): 
        # Le o arquivo e cria o dataframe com o pandas
        df = pd.read_csv(os.path.join(pasta_planilhas, arquivo))
        # Adiciona o df a lista
        dataframes.append(df)
        
print(dataframes)      
```
